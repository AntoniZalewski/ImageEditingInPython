# Praca nad następnym zdjęciem

```python
image = cv2.imread('images/SphinxCat.jpg')
```

```python
img_blur = cv2.blur(image, (15,15))
plt.imshow(img_blur[:, :, ::-1])
#porównałem i jest rzeczywiście rozmazane
```

umieścić na kopii rozmytego zdjęcia pochylony pod dowolnym kątem, czerwony prostokąt z
napisem “TOP SECRET”

```python
img_copy = img_blur.copy()

#CZERWONY KWADRACIK
rect_width = 300 * 5
rect_height = 100 * 5
rect_angle = 30
rect_x = (img_copy.shape[1] - rect_width) // 2
rect_y = (img_copy.shape[0] - rect_height) // 2
rect_center = (rect_x + rect_width // 2, rect_y + rect_height // 2)
rect = ((rect_center[0], rect_center[1]), (rect_width, rect_height), rect_angle)
box = cv2.boxPoints(rect)
box = np.int0(box)
cv2.drawContours(img_copy, [box], 0, (0, 0, 255), 3)  # Red color, thickness 3

#dodawanie napisu top secret
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 5
font_thickness = 15
text_size = cv2.getTextSize("TOP SECRET", font, font_scale, font_thickness)[0]
text_x = int(rect_center[0] - text_size[0] / 2)
text_y = int(rect_center[1] + text_size[1] / 2)
cv2.putText(img_copy, "TOP SECRET", (text_x, text_y), font, font_scale, (0, 0, 255), font_thickness)

plt.imshow(img_copy[:, :, ::-1])
plt.title('Image with Large Centered Badge')
plt.show()
```

```python
# kernel najlepiej chyba działało
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharpened_image = cv2.filter2D(image, -1, kernel)

plt.imshow(sharpened_image[:, :, ::-1])
plt.title('Wyostrzone zdjęcie')
plt.show()
```
